```python
#copy odd line of one file to other
fi1=input("Enter the source file : ")
fi2=input("Enter the destination file : ")
f1=open(fi1,"r")
f2=open(fi2,"w")
i=0
for line in f1.readlines():
    i=i+1
    if i%2==0:
        continue
    f2.write(line)
f1.close()
f2.close()
print("Odd lines copied scuccessfully from ",fi1," to ",fi2)
```

```python
#read specific row of a given csv file and print a list of str
import csv
file=input("Enter CSV file name : ")
fil=open(file,"r")
reader=csv.reader(fil)
for row in reader:
    print(row)
fil.close()
```

```python
#read specific column of a given csv file and print a list of str
import csv
files=input("Enter CSV file name : ")
with open(files,"r")as fil1:
    reader=csv.DictReader(fil1)
    print("\nDepartment Details : ")
    print("---------------")
    for row in reader:
        print("DeptId : ",row['DeptId'],"\tDepName : ",row['DeptName'])
```

```python
#write a python dictionary to a csv file.after write csv file read it, display the content
data=[
    {"Name":"Niya","Age":"21","Place":"Vadakara"},
    {"Name":"Anu","Age":"21","Place":"Nilambur"},
    {"Name":"Aswani","Age":"21","Place":"Manasheri"} 
]
with open("Std_dict.csv","w",newline='')as file:
    fieldnames=["Name","Age","Place"]
    writer=csv.DictWriter(file,fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(data)
print("Dictionary data written successfully!")
with open("Std_dict.csv","r")as file:
    reader=csv.DictReader(file)
    for row in reader:
        print(row)
```
